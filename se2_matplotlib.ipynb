# üéá Import Matplotlib Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

ÏÑ† Í∑∏ÎûòÌîÑ (Line Plot)

```python
# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
x = np.linspace(0, 10, 100)  # 0Î∂ÄÌÑ∞ 10ÍπåÏßÄ 100Í∞úÏùò Ï†ê
y = np.sin(x)  # ÏÇ¨Ïù∏ Ìï®Ïàò Í∞í

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(x, y, label="sin(x)", color="b", linestyle="--")

# Ï†úÎ™© Î∞è ÎùºÎ≤® Ï∂îÍ∞Ä
plt.title("Sinusoidal Wave")  # Ï†úÎ™©
plt.xlabel("X-axis")  # XÏ∂ï ÎùºÎ≤®
plt.ylabel("Y-axis")  # YÏ∂ï ÎùºÎ≤®
plt.legend()  # Î≤îÎ°Ä ÌëúÏãú
plt.grid(True)  # Í≤©Ïûê Ï∂îÍ∞Ä

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

ÎßâÎåÄ Í∑∏ÎûòÌîÑ (Bar Chart)

```python
# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.bar(categories, values, color=['tomato', 'skyblue', 'lightgreen', 'orange'])

# Ï†úÎ™© Î∞è ÎùºÎ≤® Ï∂îÍ∞Ä
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

ÏÇ∞Ï†êÎèÑ (Scatter Plot)

```python
# ÎûúÎç§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
plt.scatter(x, y, c=colors, cmap='viridis', alpha=0.7, edgecolors="black")

# Ï†úÎ™© Î∞è ÎùºÎ≤® Ï∂îÍ∞Ä
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Ïª¨Îü¨Î∞î Ï∂îÍ∞Ä
plt.colorbar(label="Color Intensity")

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

ÌûàÏä§ÌÜ†Í∑∏Îû® (Histogram)

```python
# ÎûúÎç§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
data = np.random.randn(1000)  # ÌèâÍ∑† 0, ÌëúÏ§ÄÌé∏Ï∞® 1Ïùò Ï†ïÍ∑úÎ∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞

# ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞
plt.hist(data, bins=30, color="skyblue", edgecolor="black")

# Ï†úÎ™© Î∞è ÎùºÎ≤® Ï∂îÍ∞Ä
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```
